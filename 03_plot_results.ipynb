```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np
import seaborn as sns

# Set the ticker symbol here. Change to any other ticker as needed.
ticker_symbol = 'AENA.MC'

# Set style
sns.set_style("ticks")
plt.rcParams['font.family'] = 'DejaVu Sans'

# 1. Load the results CSV file
results = pd.read_csv("results_lstm_lstm_dense.csv")

# 2. Convert the Date column to datetime
results['Date'] = pd.to_datetime(results['Date'], errors='coerce')

# 3. Filter the data for the chosen ticker
results_ticker = results[results['Ticker'] == ticker_symbol].copy()

# 4. Get only the test set data
test_ticker = results_ticker[results_ticker["Set"] == "Test"]

# 5. Create a single plot for the test set
fig, ax = plt.subplots(figsize=(12, 6))

# Plot test set
ax.plot(test_ticker['Date'], test_ticker['target'], label='Actual VaR', color='#1f77b4', linewidth=2)
ax.plot(test_ticker['Date'], test_ticker['prediction'], label='Predicted VaR', color='#ff7f0e', linewidth=2, alpha=0.8)
plt.xticks(rotation=45)

# Set titles and labels
ax.set_title(f'Test Set Performance - {ticker_symbol}', fontsize=16)
ax.set_ylabel('Value at Risk', fontsize=14)
ax.legend(fontsize=12)

# Adjust layout
plt.tight_layout()

# Save the figure as PNG
plt.savefig(f"{ticker_symbol}_test_set_VaR.png", dpi=300, bbox_inches='tight')

# Display the plot
plt.show()
```
